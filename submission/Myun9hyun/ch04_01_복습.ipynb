```python
from gtts import gTTS
from IPython.display import Audio
from random import choice, randint
from time import sleep

def chat():
    says = ("목아프다", "배고프다", "졸리다")
    c = choice(says)
    tts = gTTS(c, lang='ko')
    filename = "chat.mp3"
    tts.save(filename)
    audio = Audio(filename, autoplay=True)
    display(audio)

for i in range(randint(1, 5)):
    sleep(1)
    chat()
```
